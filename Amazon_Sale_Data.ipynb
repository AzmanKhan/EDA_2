Q. What is the average rating for each product category .


```python
# Convert 'rating' to numeric, setting errors='coerce' will turn non-numeric values to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Optionally, drop rows where 'rating' is NaN
df = df.dropna(subset=['rating'])

# Grouping by 'category' and calculating the mean rating
average_rating_per_category = df.groupby('category')['rating'].mean()

# Display the result
print(average_rating_per_category)
```

Q.What is the distribution of discounted prices vs. actual prices?

```python
discounted_price = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
actual_price= df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)

# Scatter plot to visualize the relationship
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x=actual_price, y=discounted_price)
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

Insights: The chart shows a strong positive correlation between actual and discounted prices, with most items clustered under ₹20,000. There are a few high-priced outliers with significant discounts, suggesting variability in discounts for expensive products.


4. How does the average discount percentage vary across categories?

5. What are the most popular product names? 


```python

# Convert 'rating_count' to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Drop rows with NaN values in 'rating_count'
df = df.dropna(subset=['rating_count'])

# Aggregate rating_count by product_name
popularity = df.groupby('product_name')['rating_count'].sum().reset_index()

# Sort by rating_count in descending order
popularity = popularity.sort_values(by='rating_count', ascending=False)

# Display the top 10 products
print(popularity.head(10))
```

6. What are the most popular product keywords? 


```python
import string
from collections import Counter

# Assuming df is your DataFrame
# Combine relevant text columns
text_data = df['about_product'].astype(str) + ' ' + df['review_content'].astype(str)

# Initialize a Counter to store word frequencies
word_counter = Counter()

# Process each text entry
for text in text_data:
    # Convert to lowercase and remove punctuation
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Split into words and update counter
    words = text.split()
    word_counter.update(words)

# Get most common keywords
most_common_keywords = word_counter.most_common(10)  # Adjust the number as needed

# Convert to DataFrame for better readability
keywords_df = pd.DataFrame(most_common_keywords, columns=['Keyword', 'Frequency'])

print(keywords_df)
```

6.Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist. 


```python
# Calculate the total popularity for each artist
total_popularity_per_artist = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False)

# Select the top 5 most popular artists
top_5_artists = total_popularity_per_artist.head(5).index

# Filter the DataFrame to include only these top 5 artists
top_5_df = df[df['Artist'].isin(top_5_artists)]

# Calculate the average popularity for each of these top 5 artists
average_popularity_per_artist = top_5_df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

pd.DataFrame(average_popularity_per_artist)
```

7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist. 


```python
# Calculate the total popularity for each artist
total_popularity_per_artist = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False)

# Select the top 5 most popular artists
top_5_artists = total_popularity_per_artist.head(5).index

# Filter the DataFrame to include only these top 5 artists
top_5_df = df[df['Artist'].isin(top_5_artists)]

# Find the most popular track for each artist
most_popular_tracks = top_5_df.loc[top_5_df.groupby('Artist')['Popularity'].idxmax()]

# Select relevant columns to display
top_5_tracks = most_popular_tracks[['Artist', 'Track Name', 'Popularity']]

pd.DataFrame(top_5_tracks)
```
